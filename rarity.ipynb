```python
import json
import pandas as pd
```

```python
df = pd.read_csv('data/kodas.csv')
```

```python
f = open('data/attribute_count.json')
attribute_dict = json.load(f)
f.close()
```

```python
df.isnull().describe()
```

```python
df.shape[0]
```

```python
for trait in attribute_dict.keys():

    trait_null_count = df[trait.lower()].isnull().describe().freq / df.shape[0]
    print (trait_null_count)

    df['{}_count'.format(trait.lower())]  = [trait_null_count if pd.isnull(x) else attribute_dict[trait][x] for x in df[trait.lower()]]   
    df['{}_rarity'.format(trait.lower())] = [trait_null_count / df.shape[0]if pd.isnull(x) else attribute_dict[trait][x] / df.shape[0] for x in df[trait.lower()]]    
```

```python
df['rarity'] = df['head_rarity'] * df['eyes_rarity'] * df['core_rarity'] * df['clothing_rarity'] * df['weapon_rarity']
```

```python
df.sort_values('rarity', ascending=True)
```

```python
df.to_csv('data/kodas_with_rarity.csv')
```
